#Here comes the hardest part to remove the ir regularities in the above phone numbers mentioned in the dataframe.


```python
df["Phone_Number"]=df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex=True)
```

```python
df
```

```python
df
```

IN order to get the number in the below format we've to make sure the entire phone number  is converted into string from int or float to make the possible.

```python
# That can be acheived by using the lambda function in python
df['Phone_Number']=df['Phone_Number'].apply(lambda x: str(x))
df
df['Phone_Number']=df['Phone_Number'].apply(lambda x:x[0:3]+'-'+x[3:6]+'-'+x[6:10] if x.isdigit() and len(x)==10 else 'NaN')
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace('NaN','', regex=True)
```

```python
df[['Street_Address','State','Zip_Code']]=df['Address'].str.split(',',n=2, expand=True )
df
```

```python
df['Do_Not_Contact']= df['Do_Not_Contact'].replace({'Yes':'Y','No':'N'})
df
```
